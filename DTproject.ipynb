```python
#importing some required libraries  
import numpy as np  
import pandas as pd  
import sklearn 
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.tree import DecisionTreeClassifier  
from sklearn.metrics import accuracy_score  
from matplotlib import pyplot as plt  
from sklearn import tree 

from google.colab import drive
drive.mount('/content/gdrive')
dataset = pd.read_csv('gdrive/My Drive/Ai labs/heart.csv')
```

```python
dataset.head() 
```

```python
x = dataset.iloc[:, 0:12] 
y = dataset.iloc[:, 13]

#priting Features and target 
print("\t\t\t\t Features")
print("____________________________________________________________________________\n")
print(x)
print("\n\t\t\t\t Target")
print("__________________________________________________________________________\n")
print(y)
```

```python
# Splitting the dataset into train and test 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 80)
```

```python
# Decision tree classifier 
clf= DecisionTreeClassifier()
#fitting the training data
clf.fit(x_train, y_train)
```

```python
# prediction on X_test (testing data)
y_prediction=clf.predict(x_test)
print(y_prediction)
```

```python
#Evaluation using Accuracy score
print("\t\t\t   Accuracy")
print("_____________________________________________________________\n")
print("The model accuracy:",metrics.accuracy_score(y_test, y_prediction))
print("\n\t\t\tConfusion Matrix")
print("_____________________________________________________________\n")
cm=np.array(confusion_matrix(y_test,y_prediction))
cm  
```

```python
print("\t\t Classification Report")
print("_______________________________________________________________")
print(classification_report(y_test, y_prediction))
print("_______________________________________________________________")
```

```python
#Tree plotting
tree.plot_tree(clf)
```
