```python
#importing some required libraries  
import numpy as np  
import pandas as pd  
import sklearn 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report, confusion_matrix 
import sklearn.metrics as metrics
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier

from google.colab import drive
drive.mount('/content/gdrive')
dataset = pd.read_csv('gdrive/My Drive/Ai labs/heart.csv')
```

```python
dataset.head()
```

```python
# Assign data from first 13 columns to X variable
x = dataset.iloc[:, 0:12]
# Assign data from 14 column to y variable
y = dataset.iloc[:, 13]
```

```python
# Splitting the dataset into train and test 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=800)
mlp.fit(x_train, y_train.values.ravel())
```

```python
# prediction on X_test (testing data)
y_prediction=mlp.predict(x_test)
print(y_prediction)
```

```python
#Evaluation using Accuracy score
print("\t\t\t   Accuracy")
print("_____________________________________________________________\n")
print("The model accuracy:",metrics.accuracy_score(y_test, y_prediction))
print("\n\t\t\tConfusion Matrix")
print("_____________________________________________________________\n")
cm=np.array(confusion_matrix(y_test,y_prediction))
cm  
```

```python
print("\t\t Classification Report")
print("_______________________________________________________________")
print(classification_report(y_test, y_prediction))
print("_______________________________________________________________")
```
